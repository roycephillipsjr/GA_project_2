```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score

%matplotlib inline
```

```python
# Reading in data
ames_housing = pd.read_csv('../datasets/clean_ames_housing_data.csv')
```

```python
ames_housing.info()
```

```python
# Filling in the Garage Type empty columns because for some reason some of the data didn't transfer over
ames_housing['Garage Type'] = ames_housing['Garage Type'].fillna('NA')
```

```python
# First Train/Test Split on all values that are 50% and above. Positive only
features = ['Overall Qual', 'gr_liv_area_and_total_bsmt', 'Exter Qual', 'Gr Liv Area',
            'Kitchen Qual', 'Garage Area','Garage Cars', 'Total Bsmt SF', '1st Flr SF',
            'Bsmt Qual', 'all_bathrooms', 'Year Built', 'Garage Finish', 'Year Remod/Add',
            'Fireplace Qu', 'Full Bath', 'TotRms AbvGrd', 'Mas Vnr Area']
X = ames_housing[features]
y = ames_housing['SalePrice']
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)

lr = LinearRegression()

cross_val_score(lr, X_train, y_train, cv=5).mean()
```

```python
# This will be my base score
lr.fit(X_train, y_train);

print('train', lr.score(X_train, y_train))

print('test', lr.score(X_test, y_test))

print('cv', cross_val_score(lr, X_train, y_train, cv=5).mean())
```

```python
# Going to do Ridge to see if there is anymore that can be done to the model
# First need to scale the data
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
)
```

```python
sc = StandardScaler()
Z_train = sc.fit_transform(X_train)
Z_test = sc.transform(X_test)
```

```python
print(f'Z_train shape is: {Z_train.shape}')
print(f'y_train shape is: {y_train.shape}')
print(f'Z_test shape is: {Z_test.shape}')
print(f'y_test shape is: {y_test.shape}')
```

```python
ols = LinearRegression()
ols.fit(Z_train, y_train)
```

```python
print(ols.score(Z_train, y_train))
print(ols.score(Z_test, y_test))
```

```python
from sklearn.linear_model import Ridge
```

```python
# Instantiate
ridge_model = Ridge(alpha=10)

# Fit.
ridge_model.fit(Z_train, y_train)

# Evaluate model using R2.
print('Train:', ridge_model.score(Z_train, y_train))
print('Test:', ridge_model.score(Z_test, y_test))
```
